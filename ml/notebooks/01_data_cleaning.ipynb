```python
# filling the missing descriptions values
df["Description"] = df["Description"].fillna("")
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"<.*?>", "", text)      # removing html
    text = re.sub(r"[^a-z\s]", "", text)   # keep letters only
    text = re.sub(r"\s+", " ", text)       # remove extra spaces
    return text.strip()
```

```python
df["clean_description"] = df["Description"].apply(clean_text)
```

```python
df["clean_genres"] = df["Genres"].str.lower()
df["clean_genres"] = df["clean_genres"].str.replace(",", " ")
```

```python
df[["Book", "clean_description", "clean_genres"]].head()
```

```python
# removing commas in num ratings
df["Num_Ratings"] = (df["Num_Ratings"].str.replace(",", "", regex=False).astype(int))
```

```python
df["Genres"].dtype
```

```python
import ast

df["Genres"] = df["Genres"].apply(ast.literal_eval)
```

```python
type(df["Genres"].iloc[0])
```

```python
df["genre_list"] = df["Genres"].apply(
    lambda x: [g.strip().lower() for g in x]
)
```

```python
df["clean_genres"] = df["genre_list"].apply(lambda x: " ".join(x))
```

```python
df[["Genres", "genre_list", "clean_genres"]].head()
# genre_list -- debug, clean_genres -- for ml model
```

```python
# data to learn from, text to compare
df["text_features"] = (df["Description"].str.lower() + " " + df["clean_genres"])
```

## ml tf-idf

```python
#vector
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf = TfidfVectorizer(
    stop_words="english",
    max_features=8000,   #metric taken for 9k books, hojana chahiye.
    ngram_range=(1, 2)
)

tfidf_matrix = tfidf.fit_transform(df["text_features"])
```

```python
tfidf_matrix.shape
```

```python
tfidf_matrix.nnz
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
import re

def normalize(text):
    return re.sub(r"\s+", " ", text.lower().strip())

df["book_norm"] = df["Book"].apply(normalize)

indices = pd.Series(df.index, index=df["book_norm"]).drop_duplicates()
```

```python
# recom function
def recommend_books(title, top_n=5):
    title_norm = normalize(title)

    # exact match
    if title_norm in indices:
        idx = indices[title_norm]

    # fallback: partial match
    else:
        matches = df[df["book_norm"].str.contains(title_norm)]

        if matches.empty:
            return "Book not found"

        idx = matches.index[0]  # best match

    sim_scores = cosine_similarity(
        tfidf_matrix[idx], tfidf_matrix
    ).flatten()

    top_idx = sim_scores.argsort()[-top_n-1:-1][::-1]

    return df.loc[top_idx, ["Book", "Author", "Avg_Rating"]]
```

```python
# test 
recommend_books("Harry Potter and the Philosopher’s Stone")
```
